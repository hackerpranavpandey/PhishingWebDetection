```python
def count(url_part,symbol):
  i=0
  for s in url_part:
    if s==symbol:
      i+=1
  return i
```

```python
def extract_tld_from_url(url1):
    if "://" in url1:
        url1 = url1.split("://")[1]
    tld = url1.split(".")[1]
    tld= tld.split("/")[0]
    return tld
```

```python
def analyze_url_components(url):
    l.append(url.count('.'))
    l.append(url.count('-'))
    l.append(url.count('_'))
    l.append(url.count('/'))
    l.append(url.count('?'))
    l.append(url.count('='))
    l.append(url.count('@'))
    l.append(url.count('&'))
    l.append(url.count('!'))
    l.append(url.count(' '))
    l.append(url.count('˜'))
    l.append(url.count(','))
    l.append(url.count('+'))
    l.append(url.count('∗'))
    l.append(url.count('#'))
    l.append(url.count('$'))
    l.append(url.count('%'))
    tld_length = len(extract_tld_from_url(url))
    l.append(tld_length)
    l.append(len(url))
```
